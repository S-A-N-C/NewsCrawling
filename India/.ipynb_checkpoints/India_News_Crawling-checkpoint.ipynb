```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
# res = requests.get('https://navbharattimes.indiatimes.com/business/business-news/articlelist/1215713388.cms')
driver = webdriver.Chrome()
url = 'https://navbharattimes.indiatimes.com/business/business-news/articlelist/1215713388.cms'
driver.get(url)
driver.maximize_window()
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

time.sleep(28) # 이유는 동적으로 기사들이 생성이 되어서 스크롤을 다 내리고 나서 생길때까지 기다려야한다. 
```

```python
# soup = BeautifulSoup(res.content, 'html.parser')
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
def news_link(tag, class_name):
    div_link = soup.find_all(tag, {"class":class_name})

    cell_line = []

    for d in div_link:
        links = d.find_all('a',href=True)
        for i in links:
            href = i.attrs['href']
            cell_line.append(href)
    
    return list(set(cell_line))          # img와 text에 href가 담겨있어서 2개씩 가져와버림 
```

```python
links_list = []
links_list.append(news_link('div','row undefined'))
links_list.append(news_link('div','row default'))
```

```python
print(links_list)
```

```python
links_list = sum(links_list, [])
```

```python
links_list = list(set(links_list))     # div 2개의 영역에서 기사를 긁어 왔기에 중복 기사가 있는지 체크
```

```python
import time
import pyautogui
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
def article_text(url):
    
    Title = ''
    Time = ''
    Contents = ''
    
    driver = webdriver.Chrome()
    driver.get(url)
    driver.maximize_window()


    pyautogui.hotkey('shift','F10')
    time.sleep(1)
    for i in range(8):
        pyautogui.hotkey('down')
    pyautogui.hotkey('enter') 
    
    time.sleep(1)
    driver.execute_script("window.scrollTo(0, window.scrollY + 200);")
    time.sleep(0.25)
    driver.execute_script("window.scrollTo(0, window.scrollY + 800);")
    time.sleep(0.25)
    
    driver.execute_script("window.scrollTo(0, window.scrollY + 200);")
    time.sleep(0.25)
    
    driver.execute_script("window.scrollTo(0, window.scrollY + 1000);")
    
    time.sleep(2)

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    article_list = []
    try:


        story = soup.find('div', {'class':'story-article'})
        h1_title = story.find('h1')
        Title = h1_title.get_text()
    except:
        pass
    
    try:

        div_time = soup.find('div',{'class':'source'})
        Time = soup.find('span',{'class':'time'}).get_text()
    except:
        pass
        
    try:
        article = story.find('article',{'class':'story-content'})
        text = article.find('font').get_text()

        p_text = article.find_all('p')
        
        if p_text[0].get_text() == 'This is a modal window.':
            p_text = article.find_all('font')
            Contents = " ".join([text, " ".join([i.get_text() for i in p_text])] ) 

        else:
            Contents  = " ".join([text, " ".join([i.get_text() for i in p_text])] ) 
    except:
        pass
        
    driver.close()
    data = pd.DataFrame({'URL':[url],'Title':[Title], 'Time' : [Time], 'Contents': [Contents]})
    return data
```

```python
from tqdm import tqdm
```

```python
df = pd.DataFrame({'URL':[], 'Title':[], 'Time' : [], 'Contents': []})
fail_df = pd.DataFrame({'URL':[]})

for u in tqdm(links_list[:120]):
    try:
        df_a = article_text(u)
        
        empty = []
        if df_a['Title'][0] == '':
            empty.append('제목')
        if df_a['Time'][0] == '':
            empty.append('시간')
        if df_a['Contents'][0] == '':
            empty.append('내용')
        
        if len(empty) != 0:
            empty = ", ".join(empty)
            raise Exception('{}의 값을 긁어오지 못했습니다.'.format(empty))
            
        df = df.append(df_a, ignore_index = True)
        
    except Exception as e:
        fail_df = fail_df.append({'URL':u, 'Error':e}, ignore_index = True)
        
    
```
